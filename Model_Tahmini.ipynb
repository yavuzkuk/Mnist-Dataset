```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import mnist

# MNIST veri setini yükleme
(train_img,train_label), (test_img, test_label) = mnist.load_data()

# Giriş görüntülerinin şeklini yeniden boyutlandırma ve normalizasyonu
train_img = train_img.reshape(train_img.shape[0], 28, 28, 1)
train_img = train_img.astype('float32') / 255.0
test_img = test_img.reshape(test_img.shape[0], 28, 28, 1)
test_img = test_img.astype('float32') / 255.0

# Etiketlerin kategorik hale getirilmesi
train_label = tf.keras.utils.to_categorical(train_label)
test_label = tf.keras.utils.to_categorical(test_label)

# Model oluşturma
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(), 
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

# Modeli derleme
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Modeli eğitme
model.fit(train_img, train_label, epochs=5, batch_size=32, validation_data=(test_img,test_label))
```

```python
# Test veri setinde tahmin yapma
predictions = model.predict(test_img)

j = 1

# Tahmin edilen sayıları ekrana yazdırma
for i in range(len(test_img)):
    true_value = np.argmax(test_label[i])
    predicted_value = np.argmax(predictions[i])
    if true_value != predicted_value:
        print("Gerçek sayi: ",true_value)
        print("Tahmin edilen sayi: ",predicted_value)
        print("Sayi: ",j)
        j = j+1
        

```
